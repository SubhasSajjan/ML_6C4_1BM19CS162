```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/content/Iris (1).csv')
feat = df.drop(['Species' , 'Id'], axis = 1)
#print(feat)
feat = df.iloc[:, [1, 2, 3, 4]].values
feat
```

```python
from sklearn.cluster import KMeans

wcss = []

for  i in range(1 , 50):
  kmeans = KMeans(n_clusters = i , init ='k-means++' , max_iter = 300 , n_init= 10 ,random_state = 0)
  kmeans.fit(feat)
  wcss.append(kmeans.inertia_)

plt.plot(range(1, 50) , wcss)
plt.title('The elbow method:')
plt.xlabel("Number of clusters")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3 , init ='k-means++' , max_iter = 300 , n_init= 10 ,random_state = 0)
y_kmeans = kmeans.fit_predict(feat)
```

```python
plt.scatter(feat[y_kmeans == 0, 0], feat[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(feat[y_kmeans == 1, 0], feat[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(feat[y_kmeans == 2, 0], feat[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
